The socket library in Python facilitates low-level network communication using the TCP/IP protocol. It serves as the foundation for building networking applications in Python, allowing for the creation of both servers and clients.

Usage:
Communication: Provides mechanisms for machines to communicate over networks.
Protocols: Enables developers to implement or customize networking protocols.


```python
import socket
import threading

def server():
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.bind(('localhost', 12347))
    s.listen(1)
    c, addr = s.accept()
    data = c.recv(1024)
    print("Server received:", data.decode())
    c.sendall(b'connected')
    c.close()

def client():
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.connect(('localhost', 12347))
    s.sendall(b'Hello--!')
    data = s.recv(1024)
    print("Client received:", data.decode())
    s.close()

# Running server and client concurrently in the notebook
threading.Thread(target=server).start()
threading.Thread(target=client).start()
```

Multithreading:
Multithreading is a technique in which multiple threads in a process execute tasks concurrently. A thread is the smallest unit of CPU execution. While multithreaded applications share the same memory space, each thread has its own set of local variables, program counter, and call stack. Multithreading can lead to better utilization of CPU resources, especially when tasks involve waiting, such as I/O operations. However, care must be taken to manage race conditions, deadlocks, and other concurrency-related issues.

```python
import threading
import math

def factorial(num):
    result = 1
    for i in range(1, num + 1):
        result *= i
    print(f"Factorial of {num} is {result}")

def square_root(num):
    result = math.sqrt(num)
    print(f"Square root of {num} is {result:.2f}")

# Create threads for the factorial and square root calculations
thread1 = threading.Thread(target=factorial, args=(21,))
thread2 = threading.Thread(target=square_root, args=(91,))

# Start the threads
thread1.start()
thread2.start()

# Wait for both threads to complete
thread1.join()
thread2.join()

print("Both tasks completed.")
```

Differences between Symmetric and Asymmetric Encryption:
Key Usage:

Symmetric: Single key for encryption & decryption.
Asymmetric: Public key for encryption, private key for decryption.
Speed:

Symmetric: Generally faster.
Asymmetric: Slower due to complex algorithms.
Key Distribution:

Symmetric: Challenge in securely sharing the key.
Asymmetric: Public key can be openly shared, simplifying distribution.
Preference:
Asymmetric encryption offers an advantage in key distribution, making it more suitable for secure communication between unfamiliar parties. However, the choice depends on the specific use case, and often systems combine both types for optimized security.

AES-256 is symmetric encryption. It uses the same key for both encryption and decryption.
